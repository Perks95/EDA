Write a Pandas program to create a side-by-side bar plot of the diamonds DataFrame.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the mean price for each new_cut category
x = diamonds.groupby("new_cut")["new_price"].mean().reset_index()

# Plotting the side-by-side bar plot
sns.barplot(x="new_cut", y="new_price", data=x)
plt.xlabel("New Cut")
plt.ylabel("Mean Price")
plt.title("Side-by-Side Bar Plot of Diamond Prices")
plt.show()
```
